```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
import tensorflow as tf
from scipy.stats import linregress
pd.set_option('display.max_columns', None)
```

```python
# Read CSV file directly without changing directory
file_path = '/content/drive/My Drive/Colab Notebooks/Resources/diabetes_prediction_dataset.csv'
diabetes_df = pd.read_csv(file_path)
diabetes_df.head()
```

```python
# Take a look at the dataset
diabetes_df.describe()
```

```python
# Check if there are any null values
diabetes_df.isnull().sum()
```

```python
# Check the dtypes of the dataset. Looks like all values are set as float64
diabetes_df.info()
```

```python
# See how many of each case there are
diabetes_count = diabetes_df.groupby('diabetes')['diabetes'].count()
diabetes_count
```

```python
# Calculate percent of total with and without diabetes
total_count = diabetes_df['diabetes'].count()
diabetes_percent = round(diabetes_df.groupby('diabetes')['diabetes'].count()/total_count * 100, 2)
diabetes_percent
```

```python
# Plot the results to see the spread of the data
diabetes_count.plot(kind="bar", width=.75)
plt.xticks(rotation=0, ha='center')
plt.xlabel("Has Diabetes?")
plt.ylabel("Total Responses")
plt.title("No Diabetes (0) vs Diabetes (1)")

plt.tight_layout()
```

```python
# Check gender column for value counts
diabetes_df['gender'].value_counts()
```

```python
# Remove 'other' from gender column because it doesn't help us predict anything
diabetes_df = diabetes_df[diabetes_df['gender'] != 'Other']
diabetes_df['gender'].value_counts()
```

```python
# Check diabetes split between gender
diabetes_gender = diabetes_df.groupby(['diabetes', 'gender'])['diabetes'].count()
diabetes_gender
```

```python
# Bar Chart showing the breakdown between gender
diabetes_gender.plot(kind='bar')

plt.xlabel('Diabetes per Gender')
plt.ylabel('Count')
plt.title

plt.show()
```

```python
# Plot some histograms to get an idea of the distribution of the data
diabetes_df.hist(figsize=(12, 8))
plt.show()
```

```python
# Generate a heatmap to see the correlation between variables
# Need to drop columns that aren't float
drop_columns = ['gender', 'smoking_history']
corr_df = diabetes_df.drop(columns=drop_columns)
corr = corr_df.corr()
sns.heatmap(corr)
```
